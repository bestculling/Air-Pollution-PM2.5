```python
pollution_df = df[['pm25', 'pm10', 'o3', 'co', 'no2', 'so2']].copy()
pollution_df
```

```python
linear = pollution_df.interpolate(method = 'linear')
```

```python
plt.figure(figsize=(12, 6))
plt.plot(pollution_df.index, pollution_df['pm25'], label='Original Data', color='blue')
plt.plot(linear.index, linear['pm25'], label='Linear Interpolation', linestyle='dotted', color='red')
plt.xlabel('Time')
plt.ylabel('PM2.5')
plt.title('Time Series of PM2.5 with Original Data (Blue) and Linear Interpolation (Red)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# à¸„à¸³à¸™à¸§à¸“à¸„à¹ˆà¸² Q1 à¹à¸¥à¸° Q3
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

# à¸„à¸³à¸™à¸§à¸“ Interquartile Range (IQR)
IQR = Q3 - Q1

# à¸à¸³à¸«à¸™à¸”à¸‚à¸­à¸šà¹€à¸‚à¸•à¸ªà¸³à¸«à¸£à¸±à¸šà¸à¸²à¸£à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸š outlier à¹‚à¸”à¸¢à¹ƒà¸Šà¹‰ IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# à¸•à¸£à¸§à¸ˆà¸ªà¸­à¸šà¹à¸¥à¸°à¸£à¸°à¸šà¸¸ outlier à¹‚à¸”à¸¢à¸”à¸¹à¸§à¹ˆà¸²à¸„à¹ˆà¸²à¹ƒà¸”à¸­à¸¢à¸¹à¹ˆà¸™à¸­à¸à¹€à¸«à¸™à¸·à¸­à¸‚à¸­à¸šà¹€à¸‚à¸•à¸—à¸µà¹ˆà¸à¸³à¸«à¸™à¸”à¹„à¸§à¹‰
outliers = (pollution_df < lower_bound) | (pollution_df > upper_bound)

# à¹à¸ªà¸”à¸‡à¸œà¸¥à¸¥à¸±à¸žà¸˜à¹Œ
print(outliers.any())
```

```python
fig, axes = plt.subplots(nrows=len(df.columns), figsize=(12, 18))

# à¹ƒà¸Šà¹‰à¸§à¸‡à¸§à¸™ for à¹€à¸žà¸·à¹ˆà¸­à¹à¸ªà¸”à¸‡ Outlier à¹ƒà¸™à¹à¸•à¹ˆà¸¥à¸°à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ
for i, column in enumerate(df.columns):
    sns.boxplot(x=df[column], ax=axes[i], color='skyblue')  # à¸žà¸¥à¹‡à¸­à¸• Boxplot à¸‚à¸­à¸‡à¸‚à¹‰à¸­à¸¡à¸¹à¸¥à¹ƒà¸™à¹à¸•à¹ˆà¸¥à¸°à¸„à¸­à¸¥à¸±à¸¡à¸™à¹Œ
    axes[i].set_title(f'Boxplot of {column}', fontsize=14)  # à¸•à¸±à¹‰à¸‡à¸Šà¸·à¹ˆà¸­à¸à¸£à¸²à¸Ÿ
    axes[i].set_xlabel('')  # à¹„à¸¡à¹ˆà¹à¸ªà¸”à¸‡à¸Šà¸·à¹ˆà¸­à¹à¸à¸™ x à¹€à¸™à¸·à¹ˆà¸­à¸‡à¸ˆà¸²à¸à¹€à¸£à¸²à¹ƒà¸Šà¹‰ Boxplot
    axes[i].set_ylabel(column, fontsize=12)  # à¸•à¸±à¹‰à¸‡à¸Šà¸·à¹ˆà¸­à¹à¸à¸™ y

plt.tight_layout()
plt.show()
```
