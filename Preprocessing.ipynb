```python
pollution_df = df[['pm25', 'pm10', 'o3', 'co', 'no2', 'so2']].copy()
pollution_df
```

```python
linear = pollution_df.interpolate(method = 'linear')
```

```python
df = linear.dropna(axis=0).reset_index(drop=True)
df
```

```python
df.isnull().sum()
```

```python
# คำนวณค่า Q1 และ Q3
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)

# คำนวณ Interquartile Range (IQR)
IQR = Q3 - Q1

# กำหนดขอบเขตสำหรับการตรวจสอบ outlier โดยใช้ IQR
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# ตรวจสอบและระบุ outlier โดยดูว่าค่าใดอยู่นอกเหนือขอบเขตที่กำหนดไว้
outliers = (pollution_df < lower_bound) | (pollution_df > upper_bound)

# แสดงผลลัพธ์
print(outliers.any())
```

```python
fig, axes = plt.subplots(nrows=len(df.columns), figsize=(12, 18))

# ใช้วงวน for เพื่อแสดง Outlier ในแต่ละคอลัมน์
for i, column in enumerate(df.columns):
    sns.boxplot(x=df[column], ax=axes[i], color='skyblue')  # พล็อต Boxplot ของข้อมูลในแต่ละคอลัมน์
    axes[i].set_title(f'Boxplot of {column}', fontsize=14)  # ตั้งชื่อกราฟ
    axes[i].set_xlabel('')  # ไม่แสดงชื่อแกน x เนื่องจากเราใช้ Boxplot
    axes[i].set_ylabel(column, fontsize=12)  # ตั้งชื่อแกน y

plt.tight_layout()
plt.show()
```
